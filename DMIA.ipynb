```
import string

def get_pwd_length(data):
    data['Length'] = pd.Series([len(i) for i in data['Password']])
    return data

def digits_in_word(word):
    ans = sum([i.isdigit() for i in word])
    return ans
    

def get_digits_num(data):
    data['Digits'] = data['Password'].apply(digits_in_word)
    return data

def get_letters_num(data):
    data['Chars'] = data['Length'] - data['Digits']
    return data

def count_all(data):
    for i in range(len(data)):
        print(i)
        pwd = data.loc[i, 'Password']
        for c in pwd:
            data.loc[i, c] += 1
    return data

def expand_this(data):
    s = set()
    [s.update(set(i)) for i in data['Password']]
    d = pd.DataFrame({str(i): 0 for i in s})
    data[d] = pd.DataFrame([[0]*len(d)], index=data.index)
    return data

def process_data(df):
    df = get_pwd_length(df)
    df = get_digits_num(df)
    df = get_letters_num(df)
    df = expand_this(df)
    df = count_all(df)
    return df
  
def normilize(data):
    data['Times'] = data['Times'].apply(lambda x: math.log(x))
    return data
    
```
